```python
#zadanie 2, count_words z codered academy, też teraz nie mam mózgu na to

def count_words(text):
    
    word_counter = 1
    #decided to add first word already in the variable reset
    
    for i in text:
        if i == " ":
            word_counter = word_counter + 1
    print(word_counter)
```

```python
#zadanie 4 - 5 min


def vowels():
    vowels = ['a','e','i','o','u','y'] #list o vovels
    
    string = input("please input string: ")
    count_vowels = 0
    
    for char in string:
        if char in vowels:
            count_vowels = count_vowels + 1
            
    print(count_vowels)
    
```

```python
#Zadanie 3 - zamysł jest, żeby kod szedł przez słowo dopóki nie napotka spacji i dodawał literki do listy,
#na koniec słowa, żeby sprawdził czy długość listy (liczba elementów) jest taka sama jak liczba elementów tej samej
#listy ale kiedy ona pokazuje tylko elementy unikatowe (set niby ma tak działać)



def duplicate_letters():
    
    text = 'Banana' #input("please input string: ")
    
    for letter in text:
        letters = []
        
        if letter != " ": #now we are in a word
            print(letter)
            letters.append(letter)  
            print(letters)

        number_of_values = len(letters)
        number_of_unique_values = len(set(letters))
        
        if number_of_values != number_of_unique_values:
            return True
        else:
            return False

    
```
